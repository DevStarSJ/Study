### 5.2 Argument

별표(\*)를 추가하면 함수는 여러 개의 매개변수를 받을 수 있다.

````python
def printa(x, y, *args):
    print "args: ", args,
    print type(args)
    print x, y,
    for item in args:
        print item,
````

### 5.3 Keyword argument

함수 인수를 전달할 때 각 매개변수의 이름과 값을 직접 지정할 수도 있다. 이러한 인수를 키워드 인수(keyword argument)라고 한다.

````python
def foo(x=10, greeting="hello", **kwargs):
    print(kwargs)
    print(type(kwargs))
    print(kwargs.get('you'))
    print(greeting, x)
````

````python
foo(you="Jimmy")
dir(foo)
````
